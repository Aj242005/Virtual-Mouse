```python
import cv2
import mediapipe as mp
import utility_functions as util
import pyautogui
```

```python
screen_width,screen_height = pyautogui.size()
pyautogui.FAILSAFE = False
print(dir(util))
hands_landmarks  = mp.solutions.hands
hands = hands_landmarks.Hands(static_image_mode = False,model_complexity=1,min_detection_confidence=0.7,min_tracking_confidence=0.7,max_num_hands=1)
```

```python
def find_finger_tip(processed):
    if processed.multi_hand_landmarks:
        hands_landmarks_pointer = processed.multi_hand_landmarks[0]
        return hands_landmarks_pointer.landmark[hands_landmarks.HandLandmark.INDEX_FINGER_TIP]
    return None
```

```python
def move_mouse(wrist):
    if wrist is not None:
        x = int(wrist.x *screen_width)
        y = int(wrist.y * screen_height)
        pyautogui.moveTo(x,y)
```

```python
def detecting_gestures(landmarks_list,processed, frame):
    if len(landmarks_list) >= 21:
        index_finger_tip = find_finger_tip(processed)

        if util.get_angle(landmarks_list[5],landmarks_list[6],landmarks_list[8]) > 90:
            move_mouse(index_finger_tip)
            print(index_finger_tip)
```

```python
def main():
    
    video = cv2.VideoCapture(0)
    draw = mp.solutions.drawing_utils
    
    try:
        while(video.isOpened()):
            ret , frame = video.read()


            if not ret:
                break
            frame = cv2.flip(frame,1)
            frameRGB = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            processed = hands.process(frameRGB)

            landmarks_list = []
            if processed.multi_hand_landmarks:
                hands_landmarks_pointer = processed.multi_hand_landmarks[0]
                draw.draw_landmarks(frame,hands_landmarks_pointer, hands_landmarks.HAND_CONNECTIONS)

                for lm in hands_landmarks_pointer.landmark:
                    landmarks_list.append((lm.x,lm.y))
                detecting_gestures(landmarks_list,processed, frame)
            cv2.imshow("ft. Sonaaa Mukhda",frame)
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break
    finally:
        video.release()
        cv2.destroyAllWindows()

if __name__ == "__main__":
    main()
```
